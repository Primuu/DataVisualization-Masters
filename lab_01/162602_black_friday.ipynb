Detection of outliers using the IQR (Interquartile range) method

```python
p0 = df.Purchase.min()
p100 = df.Purchase.max()
q1 = df.Purchase.quantile(0.25)
q2 = df.Purchase.quantile(0.5)
q3 = df.Purchase.quantile(0.75)
iqr = q3 - q1
```

```python
lc = q1 - 1.5 * iqr  # lower cutoff
uc = q3 + 1.5 * iqr  # upper cutoff
```

```python
lc
```

```python
uc
```

```python
print(f"p0 = {p0}, p100 = {p100}, lc = {lc}, uc = {uc}")
```

If lc < p0 → There are NO outliers on the bottom side
If uc > p100 → There are NO outliers on the higher side

lc < p0, so there are no outliers on the bottom side (-3523.5 <  12).
But uc < p100 (21400.5 < 23961), so there are outliers on the higher side.
A visual representation of an outlier is provided by a box plot

```python
df.Purchase.plot(kind='box')
```

Outlier treatment

```python
df.Purchase.clip(upper=uc, inplace=True)
df.Purchase.plot(kind='box')
```
