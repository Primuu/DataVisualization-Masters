Test Chi Square class implementation

```python
class ChiSquare:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None  # P-value
        self.chi2 = None  # Chi Test Statistic
        self.dof = None

        self.dfObserved = None
        self.dfExpected = None

    def _print_chisquare_result(self, colX, alpha):
        if self.p < alpha:
            result = f'{colX} is IMPORTANT for Prediction'
        else:
            result = f'{colX} is NOT important for predictor. (Discard {colX} from model).'

        print(result)

    def TestIndependence(self, colX, colY, alpha=0.25):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        self.dfObserved = pd.crosstab(X, Y)
        chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
        self.p = p
        self.chi2 = chi2
        self.dof = dof

        self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index=self.dfObserved.index)

        self._print_chisquare_result(colX, alpha)


# Initialize ChiSquare Class
cT = ChiSquare(df_train)

# Feature Selection

important_vars = []

for var in df_train.columns:
    if var != 'Basin':
        cT.TestIndependence(colX=var, colY='Basin')
        if cT.p < 0.25:
            important_vars.append(var)
            
df_train_reduced = df_train[important_vars + ['Basin']]
```

Removing outliers

```python
outlier_list = []

df_train_reduced.reset_index(drop=True, inplace=True)

for i in df_train_reduced.columns:
    q1 = np.percentile(df_train_reduced.loc[:, i], 25)
    q3 = np.percentile(df_train_reduced.loc[:, i], 75)
    step = 1.5 * (q3 - q1)
    print(f'Data points considered outliers: for the feature {i}')
    
    outliers_rows = df_train_reduced.loc[~((df_train_reduced[i] >= q1 - step) & (df_train_reduced[i] <= q3 + step)), :]
    outlier_list.append(list(outliers_rows.index))
    outliers = list(itertools.chain.from_iterable(outlier_list))
    
uniq_outlier = list(set(outliers))
```

Duplicate entries removal

```python
dup_outliers = list(set([x for x in outliers if outliers.count(x) > 1]))
print(f'Outliers list:\n{uniq_outlier}')
print(f'\nLength of outliers list: {len(uniq_outlier)}')
print(f'Duplicates list:\n{dup_outliers}')
print(f'\nLength of duplicates list: {len(dup_outliers)}')
```

```python
df_train_reduced.reset_index(drop=True, inplace=True)
real_data = df_train_reduced.drop(df_train_reduced.index[dup_outliers]).reset_index(drop=True)
```
