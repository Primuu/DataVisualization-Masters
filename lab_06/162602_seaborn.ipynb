# Distribution plot (Distplot)
### Titanic

```python
sns.set_theme(palette='bright')
sns.distplot(titanic.age, rug=True, fit=stats.gausshyper)
```

### Toxins

```python
sns.set_theme(palette='husl')
sns.distplot(toxins.TN, rug=True, fit=stats.gausshyper)
```

# Joint Plot (Joinplot)
### Titanic

```python
sns.set_theme(palette='bright')
sns.jointplot(x=titanic.age, y=titanic.fare)
```

### Toxins

```python
sns.set_theme(palette='husl')
sns.jointplot(x=toxins.TN, y=toxins.Chlorophyla)
```

# Joint Plot with hexagons (Hex Joinplot)
### Titanic

```python
with sns.axes_style('white'):
    sns.set_theme(palette='bright')
    sns.jointplot(x=titanic.age,
                  y=titanic.fare,
                  kind='hex',
                  color='b')
```

### Toxins

```python
with sns.axes_style('white'):
    sns.jointplot(x=toxins.TN,
                  y=toxins.Chlorophyla,
                  kind='hex',
                  color='r')
```

# Density Plot
### Titanic

```python
sns.set_theme(palette='bright')
sns.jointplot(x=titanic.age, y=titanic.fare, kind='kde', fill=True)
```

### Toxins

```python
sns.set_theme(palette='husl')
sns.jointplot(x=toxins.TN, y=toxins.Chlorophyla, kind='kde', fill=True)
```

### Titanic - customized

```python
sns.set_theme(palette='bright')
titanic_jp = sns.jointplot(x=titanic.age, y=titanic.fare, kind='kde', color='b', fill=True)
titanic_jp.plot_joint(plt.scatter, c='black', s=30, linewidth=0.5, marker='+')
titanic_jp.ax_joint.collections[0].set_alpha(0.9)
```

### Toxins - customized

```python
sns.set_theme(palette='husl')
toxins_jp = sns.jointplot(x=toxins.TN, y=toxins.Chlorophyla, kind='kde', color='r', fill=True)
toxins_jp.plot_joint(plt.scatter, c='black', s=30, linewidth=0.5, marker='o')
toxins_jp.ax_joint.collections[0].set_alpha(0.9)
```

# Query (filter + regression)
### Titanic

```python
sns.set_theme(palette='bright')
sns.lmplot(x='age', y='fare', data=titanic.query("`class` == 'First'"), ci=None, scatter_kws={'s': 80})
```

### Toxins

```python
sns.set_theme(palette='husl')
sns.lmplot(x='TN', y='Chlorophyla', data=toxins.query("TTC > 10"), ci=None, scatter_kws={'s': 80})
```

## With parameter adjustment
### Titanic

```python
sns.set_theme(palette='bright')
sns.lmplot(x='age', y='fare', data=titanic.query("`class` == 'First'"), order=10, ci=None, scatter_kws={'s': 80})
```

### Toxins

```python
sns.set_theme(palette='husl')
sns.lmplot(x='TN', y='Chlorophyla', data=toxins.query("TTC > 10"), order=5, ci=None, scatter_kws={'s': 80})
```

# Non-linear relationship with additional data (customized)
### Titanic
